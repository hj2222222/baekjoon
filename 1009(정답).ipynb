```python
t = int(input())
z = []
for i in range(t):
  a, b = map(int, input().split(' '))
  a = a % 10
  c = a
  for _ in range(b-1):
    c *= a
    c = c % 10
  z.append(c)
for q in range(len(z)):
  print(z[q])

#5
#1 6
#3 7
#6 2
#7 100
#9 635
```
