```python
#1145
#자연수 5개가 주어질때 적어도 3개가 나누어지는 가장 작은 자연수(시간초과)
a = list(map(int,input().split(' ')))
i = 1
z = []
x = []
while True:
  for j in range(len(a)):
    z.append(a[j]*i)
  print(z)
  b = set(z)
  print(b)
  for k in range(len(b)):
    b = list(b)
    x.append(z.count(b[k]))
    if z.count(b[k]) == 3:
      print(b[k])
  if 3 in x:
    break
    b = set(b)
  i += 1
```

```python
#1145 (정답)
a = list(map(int,input().split(' ')))
t = min(a)
while True:
  z = []
  for i in range(len(a)):
    z.append(t%a[i])
  if z.count(0) >= 3:
    break
  t += 1
print(t)
```

```python
#1546 (정답)
t = int(input())
a = list(map(int,input().split(' ')))
b = max(a)
for i in range(len(a)):
  a[i] = a[i]/b*100
print(sum(a)/len(a))
```

```python
#1834(정답)
t = int(input())
count = 0
j = (t+1)
k = (t+1)
for _ in range(t-1):
  count += j
  j += k
print(count)
```

```python
#8595 (정답)
t = int(input())
a = input()
z = []
sum = 0
b = str(0)
for i in range(len(a)):
  if a[i].isdigit():
    b = b+str(a[i])
    if i == (len(a)-1):
      z.append(b) 
  else:
    z.append(b)
    b = str(0)
for i in range(len(z)):
  sum += int(z[i])
print(z)
print(sum)
```
