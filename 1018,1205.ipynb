```python
#1018
n,m = map(int,input().split(' ')) # array 활용??
z = []
for i in range(n):
  lst = list(input())
  z.append(lst)
print(z)
y = z.copy()
if z[0][0] == 'B':
  for i in range(n):
    for j in range(1,m+1):
      y[2*i-1][2*j-1] = 'B'
      y[2*i][2*j] = 'W'
elif z[0][0] == 'W':
  for i in range(n):
    for j in range(1,m+1):
      y[2*i-1][2*j-1] = 'W'
      y[2*i][2*j] = 'B'
print((z == y).conut(False))
```

```python
#1205
n,s,p = map(int,input().split(' '))
if n > 0:
  lst = list(map(int,input().split(' ')))
  if s < lst[-1]:
    print(-1)
  elif s == lst[-1]:
    if len(lst) < p:
      lst.append(s)
      print(lst.index(s)+1)
    elif len(lst) >= p:
      print(-1)
  elif s > lst[-1]:
    if len(lst) < p:
      lst.append(s)
      lst.sort(reverse=True)
      print(lst.index(s)+1)
    elif len(lst) >= p:
      del lst[-1]
      lst.append(s)
      lst.sort(reverse=True)
      print(lst.index(s)+1)
else:
  print(1)
```
