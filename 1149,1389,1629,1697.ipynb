```python
#1697
a,b = map(int,input().split(' '))
cnt = 0
while True:
  if a == b:
    print(cnt)
    break 
  else:
    if (b+1) >= 2*a:
      a *= 2
      cnt += 1
    #elif b > 2*a:
      #a *= 2
      cnt += 1
    else:
      if b % 2 ==0:
        c = a-(a-b/2)
        a -= c
        cnt += c
      else:
        c = a-(a-(b+1)/2)
        a -= c
        cnt += c
```
