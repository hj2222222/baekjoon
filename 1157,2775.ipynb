```python
#1157
#가장 많은 알파벳이 나온것을 출력 - 문자열 입력 - 모두 소문자로 변경 - 각각의 문자를 카운터하기(모든 알파벳을 기준으로?) - 나온것들 빈도수 비교 - 결과값 출력
t = input()
t = t.lower()
z = []
al = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
for i in al:
  a = t.count(i)
  z.append(a)
if z.count(max(z)) > 1:
  print('?')
elif z.count(max(z)) == 1:
  y = al[z.index(max(z))]
  y = y.upper()
  print(y)
```

```python
#2775
#0층 1 2 3 4 1층 1 3 6 10 2층 1 4 10 20 .... 
z = []
y = []
ret=0
t = int(input())
for _ in range(t):
  a = int(input())
  b = int(input())
  z0=list(range(1,b+1))
  for i in range(1,a+1):
    for j in range(1,b+1):
      if j == 1:
        z0.append(1)
      elif j > 1:
        z = z0[b*(i-1):b*(i-1)+j]
        ret += sum(z)
        z0.append(ret)
        z = []
        ret = 0
  print(z0)
  y.append(z0[-1])
print(y)
for i in range(len(y)):
  print(y[i])
```
